```python
import pandas as pd

# read resources/2017.csv and set into a mhs_2017 DataFrame
mhs_2017 = pd.read_csv("resources/2017.csv")

# read resources/2018.csv and set into a mhs_2018 DataFrame
mhs_2018 = pd.read_csv("resources/2018.csv")

# read resources/2019.csv and set into a mhs_2019 DataFrame
mhs_2019 = pd.read_csv("resources/2019.csv")

# read resources/2020.csv and set into a mhs_2020 DataFrame
mhs_2020 = pd.read_csv("resources/2020.csv")

# read resources/2021.csv and set into a mhs_2021 DataFrame
mhs_2021 = pd.read_csv("resources/2021.csv")

# read resources/2022.csv and set into a mhs_2022 DataFrame
mhs_2022 = pd.read_csv("resources/2022.csv")

# read resources/2023.csv and set into a mhs_2023 DataFrame
mhs_2023 = pd.read_csv("resources/2023.csv")


```

```python

merged_df = pd.concat([mhs_2017, mhs_2018, mhs_2019, mhs_2020, mhs_2021, mhs_2022, mhs_2023])
display(merged_df.head())
```

```python
merged_df.columns

```

```python

# Function to rename columns
def rename_columns(column):
    return column.lower().replace(" ", "-")

# Rename the columns of merged_df using the rename() method
merged_df = merged_df.rename(columns=rename_columns)

# Display the updated DataFrame
display(merged_df.head())
```
